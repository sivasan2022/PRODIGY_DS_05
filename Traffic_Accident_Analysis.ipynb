Preprocessing the dataset

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
normalized_data=df.apply(LabelEncoder().fit_transform)
normalized_data.head()
```

```python
normalized_data.dtypes
```

Data visualization

```python
# 1. Distribution of age of the driver
plt.figure(figsize=(10, 6))
sns.histplot(df['Age_band_of_driver'], bins=50, kde=True)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Ages")
plt.show()
```

```python
# 2. Pie plot for Accident severity
plt.figure(figsize=(20, 10))
palette_color = sns.color_palette('pastel')
df1=pd.DataFrame(df['Accident_severity'].value_counts().reset_index())
plt.pie(data=df1, x='Accident_severity', labels='index', colors=palette_color, autopct='%.0f%%')
plt.title('Accident Severity')
plt.show()
```

```python
cause_severity_counts = df.groupby(['Cause_of_accident', 'Accident_severity']).size().unstack().fillna(0)

# Create a stacked bar plot
cause_severity_counts.plot(kind='bar', stacked=True)
plt.title('Accident Causes and Severity')
plt.xlabel('Cause')
plt.ylabel('Count')
```

```python
from seaborn.widgets import dark_palette
weather_severity_counts = df.groupby(['Weather_conditions', 'Accident_severity']).size().unstack().fillna(0)

# Create a stacked bar plot
weather_severity_counts.plot(kind='bar', stacked=True)
plt.title('Weather Conditions and Severity')
plt.xlabel('Weather Conditions')
plt.ylabel('Count')
```

```python
weather_severity_counts = df.groupby(['Road_surface_type', 'Accident_severity']).size().unstack().fillna(0)

# Create a stacked bar plot
weather_severity_counts.plot(kind='bar', stacked=True)
plt.title('Road conditions and Severity')
plt.xlabel('Road Conditions')
plt.ylabel('Count')
```

```python
# 5. Gender distribution of drivers
plt.figure(figsize=(6, 4))
sns.countplot(x='Sex_of_driver', data=df)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Gender Distribution of Drivers")
plt.show()
```
